```python
prefix = 'https://www.reuters.com/news/archive/'
place = 'russia'
postfix = '?view=page&page='
```

```python
max_page = 2
```

```python
def get_1page(page_num):
    time_stamps, headline, abstract, link = [],[],[],[]
    url = prefix+place+postfix+str(page_num)
#     print(url)
    response = requests.get(url,headers={'User-Agent': 'Mozilla/5.0'},verify=True)
    page= bs(response.content,'html.parser')
#     print(page.prettify())
    
#     for i in page.find_all('div', class_='news-headline-list').find_all('h3'):
    for i in page.find('div', class_='news-headline-list').find_all('h3', class_='story-title'):
        s = re.sub(' {2,}', '', i.text)
        s = re.sub(r'[\n\t\r]','',s)
        headline.append(s)
        
    for i in page.find('div', class_='news-headline-list').find_all('span', class_='timestamp'):
        #replace with today if not date given
        time_stamps.append(i.text)    

#     for i in page.find('div', class_='news-headline-list').find_all('div', class_='contributor'):
#         s = re.sub(' {2,}', '', i.text)
#         s = re.sub(r'[\n\t\r]','',s)
#         abstract.append(i.text)    
    
    # to df
    
    
#     print(f'Length of time_stamp ={len(time_stamps)},\n Length of headlines = {len(headline)},\n Length of the abstracts = {len(abstract)}')
    
#     print(time_stamps)
#     print(headline)
#     print(abstract)
    
    return headline
    
```

```python
def get_pages(max_page,snap_time=1):
    if not isinstance(max_page,int):
        max_page = int(max_page)
    
    headline = []
    for num in range(1, max_page+1):
        time.sleep(snap_time)
        # add try except
        # add progress
        headline+=get_1page(num)
    # add database
    # add pickle
    print(f'total number of articles = {len(headline)}\n')
    return headline
```
